```python
import pandas as pd
import numpy as np

loan_data = pd.read_csv('FeatureEnggToDo.csv',error_bad_lines=False,encoding = "utf-8")
loan_data.drop(['issue_d'],axis=1,inplace=True)
labels = list(loan_data.columns.values)
data_array = np.array(loan_data)

x = data_array[:, 1:data_array.shape[1]]
y_train=x[:,2]
X_train=x[:,[v for v in range(len(x[0])) if v !=2]]


import sklearn.datasets
import autosklearn.regression

X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(X_train,y_train, random_state=1)

automl = autosklearn.regression.AutoSklearnRegressor()
automl.fit(X_train, y_train)

predictions = automl.predict(X_test)
```

```python
print(predictions)
```

```python
pred = pd.DataFrame({'prediction' : predictions})
pred.to_csv('AutoSKLearn_prediction.csv',index = False, sep = ' ')
```
